```python
def add_unique(x, seen=None):
    if seen is None:
        seen = set()
        if x not in seen:
            seen.add(x)
        return list(seen)
```

```python
def moving_average(stream,w):
    result = []
    current_sum = 0

    for i in range(len(stream)):
        current_sum += stream[i]

        if i >= w:
            current_sum -= stream[i - w ]
            windows_size = min(i + 1, w)
            result.append(current_sum / windows_size)

            return result
```
